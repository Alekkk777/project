4.3-Upload Image e salvataggio file in locale
OSS: 
- @st.cache è un decoratore fornito dalla libreria Streamlit in Python, usato per memorizzare nella cache i risultati di funzioni costose in termini di tempo di esecuzione
- questa funzione per il salvataggio in locale può essere utilizzata anche per i testi o i csv

```python
#ci serve soltanto per caricare una immagine random
from PIL import Image
#ci serve per salvare in locale (os ci permette di lavorare con le componenti del nostro pc)
import os
#scriviamo una funzione per il caricamento dell'immagine
@st.cache
def load_image(image_file):
    img = Image.open(image_file)
    return img

#scriviamo una funzione per salvare il file in locale
def save_uploaded_file(uploadedfile):
    with open(os.path.join("tempDir",uploadedfile.name),"wb") as f:
        f.write(uploadedfile.getbuffer())
    return st.success("Saved :{} in tempDir".format(uploadedfile.name))

def main():
    uploadedfiles = st.file_uploader("Upload multiple Images",type=["png","jpg","jpeg"],accept_multiple_files=True))
        if uploadedfiles is not None:
            st.write(uploadedfiles)
            for imagefile in uploadedfiles:
                st.write(image_file.name)
                st.image(load_image(image_file),width=250,height=250)
                save_uploaded_file(image_file)
```

Facciamo un mini esercizio in cui chiediamo di fare upload di un file, poi lo modifichiamo e poi possono fare il download del file modificato

```python
import pandas as pd
import time
timestr=time.strftime("%Y%m%d-%H%M%S")
#funzione per caricare il file
def load_data(data):
    return pd.read_csv(data)

def main():
    #chiediamo upload file
    data_file = st.file_uploader("Upload CSV", type=["csv"])
    if data_file is not None:
        df = load_data(data_file)
        # chiediamo la modfica con un form
        with st.form("editor_form"):
            edited_df = st.data_editor(df)
            save_button = st.form_submit_button("Save Data")
        #aggiungiamo un bottone così che quando finisca le modifiche, cliccando su save_button crei un nuovo file 
        #con le modifiche(timestr ci serve per rinonimnare il file in maniera univoca)
        if save button:
            new filename = f"{data file. name} {timestr}.csv"
            final_df = edited_df.to_csv()
        #diamo il permesso di scaricare il nuovo file
            st.download_button (label="Download data as CSV",data=final_df, file_name=new_filename,mime='text/csv')
if __name__ == '__main__':
    main()
```

```python
import streamlit as st 
import streamlit.components as stc

# Utils
#Libreria per la conversione di dati in codifica base64, utilizzata per la generazione di link di download(utile per il download dei file in maniera sicura)
import base64 
#Libreria per la gestione del tempo, utilizzata per ottenere una stringa di data/ora formattata.
import time
timestr = time.strftime("%Y%m%d-%H%M%S")
import pandas as pd 



#Funzione per il download in formato txt
def text_downloader(raw_text):
# Codifica il testo in base64
    b64 = base64.b64encode(raw_text.encode()).decode()
# Genera un nome univoco per il file utilizzando la data/ora corrente
    new_filename = "new_text_file_{}_.txt".format(timestr)
# Mostra un titolo nella pagina web
    st.markdown("#### Download File ###")
# Genera un link per il download del file di testo
    href = f'<a href="data:file/txt;base64,{b64}" download="{new_filename}">Click Here!!</a>'
#mostra il link nella pagina
    st.markdown(href,unsafe_allow_html=True)

#Funzione per il download in formato txt
def csv_downloader(data):
    csvfile = data.to_csv()
    b64 = base64.b64encode(csvfile.encode()).decode()
    new_filename = "new_text_file_{}_.csv".format(timestr)
    st.markdown("#### Download File ###")
    href = f'<a href="data:file/csv;base64,{b64}" download="{new_filename}">Click Here!!</a>'
    st.markdown(href,unsafe_allow_html=True)

#Gestisce il download dei file
class FileDownloader(object):
    """docstring for FileDownloader
    >>> download = FileDownloader(data,filename,file_ext).download()

    """
#il costrutto viene chiamato quando si crea un nuovo oggetto FileDownloader(Imposta i dati da scaricare, il nome del file e l'estensione del file.)
def __init__(self, data,filename='myfile',file_ext='txt'):
    super(FileDownloader, self).__init__()
    self.data = data
    self.filename = filename
    self.file_ext = file_ext
#Il metodo download converte i dati forniti in base64 e genera un link di download.
def download(self):
    b64 = base64.b64encode(self.data.encode()).decode()
    new_filename = "{}_{}_.{}".format(self.filename,timestr,self.file_ext)
    st.markdown("#### Download File ###")
    href = f'<a href="data:file/{self.file_ext};base64,{b64}" download="{new_filename}">Click Here!!</a>'
    st.markdown(href,unsafe_allow_html=True)





def main():
my_text = st.text_area("Your Message")
st.write(my_text)
# text_downloader(my_text)
download = FileDownloader(my_text).download()


df = pd.read_csv("iris.csv")
st.dataframe(df)
# csv_downloader(df)
download = FileDownloader(df.to_csv(),file_ext='csv').download()

if __name__ == '__main__':
    main()
```
